```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
import Elevators
%matplotlib notebook
```

```python
x_max = 100
N = 10

X = np.arange(1, x_max + 1)
P1, P2 = Elevators.gen_graphing_list(N, x_max)
```

```python
fig = plt.figure()
points = fig.add_subplot(111)
points.plot(X, P1, 'r.', label = 'P1 average $\\frac{transit\\;time}{query}$')
points.plot(X, P2, 'b.', label = 'P2 average $\\frac{transit\\;time}{query}$')

points.set_xlabel('$Q_1$')
points.set_ylabel('$\\frac{transit\\;time}{query}$')
points.set_title('Change in Average $\\frac{transit\\;time}{query}$ as $P_1$ dominates $P_2$\n(N = 10, $P_1,P_2$ act random)')
points.legend(loc = 'best')

plt.savefig('P1_dom_P2_N10.pdf')
plt.savefig('P1_dom_P2_N10.png')
```

```python
x_max = 11
N = 10

X = np.arange(1, x_max + 1)
P1, P2 = Elevators.gen_sym_graphing_list(N, x_max)
```
